## Addition

```python
df['add_sepal_petal_length'] = df['sepal_length'] + df['petal_length']
df['add_sepal_petal_length']
```

```python
df.sample(3)
```

## Subtrction

```python
df['sub_sepal_petal_length'] = df['sepal_length'] - df['petal_length']
df['sub_sepal_petal_length']
```

```python
df.sample(3)
```

## Multiplication

```python
df['multiply_sepal_petal_length'] = df['sepal_length'] * df['petal_length']
df['multiply_sepal_petal_length']
```

## Division

```python
df['ratio_sepal_length_width'] = df['sepal_length'] / df['sepal_width']
df['ratio_sepal_length_width']
```

## Exponential

```python
df['sepal_length_squared'] = df['sepal_length'] ** 2
df['sepal_length_squared']
```

```python
df['sepal_length_squared'] / 2
```

## How to rename columns or rows in pandas

```python
df.columns
```

```python
df.sample(3)
```

```python
# rename columns
rename_dict = {'sepal_length': 'SL',
               'sepal_width': 'SW'}
```

```python
df.rename(rename_dict, axis='columns', inplace=True) # axis = 1 also works
```

```python
df.head()
```

```python
# rename rows
r_rows_dict = {0: '11111111'}
df.rename(r_rows_dict, axis='index', inplace=True)
```
