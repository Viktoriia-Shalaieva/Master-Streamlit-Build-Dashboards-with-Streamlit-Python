Univariate analysis

How is the sepal length distributed?

```python
for col in ['sepal_width', 'petal_length', 'petal_width', 'sepal_length']:
    plot = px.histogram(data_frame=df,
                    x=col,
                    facet_col='species',
                    title='Distribution of {}'.format(col))
    plot.show()
```

```python
# create box plot
for col in ['sepal_width', 'petal_length', 'petal_width', 'sepal_length']:
    plot = px.box(data_frame=df,
                    y=col,
                    x='species',
                    title='Spread of {}'.format(col))
    plot.show()
```
